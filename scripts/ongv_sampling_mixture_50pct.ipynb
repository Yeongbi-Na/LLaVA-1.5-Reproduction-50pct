```python
import json
import random

input_file = "/nas/datahub/llava-v1.5-instruct/llava_v1_5_mix665k.json"
output_file = "/nas/datahub/llava-v1.5-instruct/llava_v1_5_mix665k_50pct.json"
```

```python
print(f"Loading {input_file}...")
with open(input_file, "r") as f:
    data = json.load(f)
```

```python
total_len = len(data)
print(f"Total samples: {total_len}")

# 2. Random Shuffle (재현성을 위해 seed 고정)
random.seed(42) 
random.shuffle(data)

# 3. 50% Slicing
subset_len = int(total_len * 0.5)
subset_data = data[:subset_len]

print(f"Selected samples: {len(subset_data)} (50%)")

# 4. Save
print(f"Saving to {output_file}...")
with open(output_file, "w") as f:
    json.dump(subset_data, f, indent=2)

print("Done!")
```
